```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style
```

```python
df = pd.read_csv("data/train.csv")
df.head(50)
```

```python
df.info()
```

```python
print(df.describe())
```

```python
# Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['pressure'], kde=True)
plt.title("Distribution of Pressure")
plt.xlabel("Pressure (cmH2O)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,6), dpi= 80)
sns.heatmap(
    df.corr(), xticklabels=df.corr().columns,
    yticklabels=df.corr().columns, cmap='Blues', center=0, annot=True, cbar=False
)
plt.title('Correlation between features', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
